```python
import numpy as np 
import matplotlib.pyplot as plt 
```

# Voici une fonction que je souhaite approximer la solution.
Vous pouvez changer cette fonction pour trouver la solution de votre fonction.

```python
def f(x):
    return x**2 - 2
```

```python
def dichotomie_graphique(f, a, b, n_iter = 4):
    # Creation des points pour la courbe
    x = np.linspace(a - 0.5, b + 0.5, 400)
    y = f(x)

    plt.figure(figsize=(10,6))
    plt.plot(x, y, 'k-', label=r'$f(x) = x^2 - 2$', linewidth=2)
    plt.axhline(0, color='black', linewidth=1) # axe des abscisses

    # Couleurs pour les differents etapes
    colors = plt.cm.viridis(np.linspace(0, 1, n_iter))

    print(f"{'Iter.:<10'} {'a:<5'} {'b:<10'} {'m (milieu):<10'} {'f(m):<10'}")

    print("-" * 55)

    for i in range(n_iter):
        m = (a + b) / 2
        fm = f(m)

        # Affichage console
        print(f"{i+1:<10} {a:<10.4f} {b:<10.4f} {m:<10.4f} {fm:<10.4f}")

        # Dessin du segment et du point milieu
        plt.plot([a,b], [0 ,0], color = colors[i], alpha=0.3, linewidth = 5)
        plt.vlines(m, 0, fm, color = colors[i], linestyle = '--', alpha = 0.7)
        plt.scatter(m, 0, color = colors[i], zorder = 5, label = f'Iter {i+1}: m = {m:.3f}')

        # logique de la dichotomie
        if f(a) * fm < 0:
            b = m
        else:
            a = m
    # Esthetique de graphique
    plt.title("Visualisation de la methode de dichotomie")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.5)
    plt.show()
```

```python
# Execution
dichotomie_graphique(f, 1, 2, n_iter = 5)
```
